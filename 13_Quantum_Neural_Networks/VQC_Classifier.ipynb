# Quantum Neural Networks: Variational Quantum Classifier (VQC)

This notebook implements a Variational Quantum Classifier (VQC). It demonstrates how to construct a hybrid Quantum Neural Network (QNN) to classify complex, non-linear data, representing the cutting edge of Quantum Machine Learning (QML).

## QSVM vs. VQC

In a previous project, we built a Quantum Support Vector Machine (QSVM). In that model, the quantum computer simply calculated the "distances" (the kernel) between data points, and handed that math over to a classical AI to draw the boundary.

A **Variational Quantum Classifier (VQC)** is fundamentally different. It acts exactly like a classical Deep Neural Network, but its "hidden layers" are made of quantum entanglement and superposition.

## The Architecture of a QNN

A VQC operates using three distinct quantum components inside a classical optimization loop:

1. **The Feature Map (Data Encoding):** Classical data (like image pixels or financial metrics) cannot simply be "pasted" into a quantum computer. We use a parameterized circuit to rotate qubits based on the values of our data, embedding classical information into a high-dimensional quantum Hilbert space.
2. **The Ansatz (The Quantum Layers):** This is the "brain" of the QNN. It is a sequence of highly entangled quantum gates with adjustable rotation angles ($\theta$). These are the "weights and biases" of our neural network.
3. **Measurement & Optimization:** We measure the qubits to get a classification prediction (e.g., `0` for Cat, `1` for Dog). A classical optimizer (like COBYLA) calculates the loss (how wrong the prediction was) and updates the $\theta$ parameters in the Ansatz to make the network smarter for the next round.

In the code below, we will use Qiskit's Machine Learning library to train a VQC to classify a non-linear dataset, plotting the training loss over time just as we would in PyTorch or TensorFlow.
